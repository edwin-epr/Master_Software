```julia
p = plot(x1,[y1,y2, y3], label=["Jacobi" "Gauss-Seidel" "Sobrerelajación Sucesiva"], lw=[3 3 3], marker=:diamond)
title!("Métodos Iterativos Clásicos")
plot!(xaxis=("Tamaño matriz") , yaxis = ("Tiempos de ejecución (s)", :log))
savefig(p, "runtime_metodos_iterativos_clasicos")
display(p)
```
